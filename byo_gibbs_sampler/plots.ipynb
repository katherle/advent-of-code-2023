```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ex1_results = pd.read_csv("ex1_results.txt", names = ["data", "mu_grid", "lnL"], delim_whitespace=True)
mu = 10.
sigma = 1.
```

```python
plt.figure()
plt.hist(ex1_results["data"], bins = 30)
plt.xlabel('Amplitude',size=15)
plt.ylabel('Bin Count',size=15)
plt.axvline(mu,color='k',linestyle='--')
plt.axvline(mu-sigma,color='k',linestyle='--')
plt.axvline(mu+sigma,color='k',linestyle='--')
plt.title("Example 1 data")
plt.show()
```

```python
plt.figure()
plt.plot(ex1_results["mu_grid"], ex1_results["lnL"])
plt.axvline(9.974136, color = "k", linestyle = "--", linewidth = 1)
plt.xlabel(r"$\mu$")
plt.ylabel(r"$\mathcal{L}$")
plt.title("Example 1 results")
plt.show()
```

```python
ex2_results = pd.read_csv("ex2_results.txt", names = ["mu_grid", "lnL"], delim_whitespace=True)
data_mean = np.array([4.974137, 9.989951, 14.99854, 20.10530, 24.98831])
data_error = np.array([1.017731, 1.606884, 1.276801, 1.967768, 0.7675156])
nus = np.array([10., 20., 30., 40., 50.])

plt.figure()
plt.errorbar(nus, data_mean, yerr = data_error, fmt = ".", color = "k")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Example 2 data")
plt.show()
```

```python
plt.figure()
plt.plot(ex2_results["mu_grid"], ex2_results["lnL"])
plt.axvline(10.00750, ls = "--", lw = 1, color = "k")
plt.xlabel(r"$a_{\rm obs}$")
plt.ylabel(r"$\mathcal{L}$")
plt.title("Example 2 results")
plt.show()
```
